#### **Pseudocódigo:**

```plaintext
função number_of_connectives(A):
    se A é um Atom:
        retorne 0
    se A é um Not:
        retorne 1 + number_of_connectives(A.inner)
    se A é um Implies:
        retorne 1 + number_of_connectives(A.left) + number_of_connectives(A.right)
    se A é um And:
        retorne 1 + number_of_connectives(A.left) + number_of_connectives(A.right)
    se A é um Or:
        retorne 1 + number_of_connectives(A.left) + number_of_connectives(A.right)

```
---
#### **Implementação:**

```python
def number_of_connectives(formula: Formula):
    """Returns the number of connectives occurring in a formula."""
    if isinstance(formula, Atom):
        return 0  # Átomo não conta como conectivo
    elif isinstance(formula, Not):
        return 1 + number_of_connectives(formula.inner)  # 1 para a negação + conectivos da subfórmula
    elif isinstance(formula, Implies):
        return 1 + number_of_connectives(formula.left) + number_of_connectives(formula.right)  # 1 para implica
    elif isinstance(formula, And):
        return 1 + number_of_connectives(formula.left) + number_of_connectives(formula.right)  # 1 para and
    elif isinstance(formula, Or):
        return 1 + number_of_connectives(formula.left) + number_of_connectives(formula.right)  # 1 para or
    else:
        return 0  # Para fórmulas não reconhecidas

```

### **Item A**
#### **Passos para a Conversão**

**1. Identificar a Estrutura da Expressão:**  
A expressão é `¬(p → ¬q)`.  
* Há uma negação externa (¬) e uma implicação interna (→).  

**2. Identificar os Operadores e Operandos:**

| **Operadores**       | **Operandos da Implicação**           |
|:--------------------:|:-------------------------------------:|
| Operador externo: ¬  | Primeiro operando: p                  |
| Operador interno: →  | Segundo operando: ¬q (q é negado!)|


**3. Converter a Implicação para Notação Polonesa:**  
* A implicação (→) é representada antes de seus operandos:  
    * Portanto, a implicação se torna `→ p ¬q.`  

**4. Colocar o Operador Externo Primeiro:**  
* Agora, colocamos o operador de negação (¬) antes da implicação:
    * Isso resulta em `¬ → p ¬q`.

#### **Resultado Final:**  
Assim, a expressão `¬(p → ¬q)` em notação polonesa é: `¬ → p ¬ q`
